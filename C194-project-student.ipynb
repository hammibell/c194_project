# Task 1 - Plot a line graph to show the difference between heart rate of smokers and non smokers

```python
#Filter and make a new dataframe for non smokers

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

smoke = pd.read_csv("framingham.csv")

df = smoke.dropna()


non_smokers = df.loc[df["currentSmoker"] == 0]
non_smokers
```

```python
#Group by age column and find average heart rate at different age

group = non_smokers.groupby("heartRate")["age"].mean().reset_index()
group
```

```python
#Filter and make a new dataframe for smokers

smokers = df.loc[df["currentSmoker"] == 1]
smokers
```

```python
#Group by age column and find average heart rate at different age


different_group = smokers.groupby("heartRate")["age"].mean().reset_index()
different_group
```

```python
#Plot a line graph to show the heart rate of smokers vs non smokers

plt.figure(figsize = (16, 9))

plt.plot(group["heartRate"], group["age"], label = "Non Smokers", color = "blue")
plt.legend()

plt.plot(different_group["heartRate"], different_group["age"], label = "Smokers", color = "red")
plt.legend()


plt.show()
```

# Task 2 - Which age group have high chances of having coronary heart disease in 10 years

```python
#Read the csv

import pandas as pd
import matplotlib.pyplot as plt

heart = pd.read_csv("framingham.csv")

df = heart.dropna()
df



#Filter and make a new dataframe for those who has chances of having coronary heart disease in 10 years

ten_years = df.loc[df["TenYearCHD"] == 1]
ten_years
```

```python
#Group by age column and count the rows of TenYearCHD column

group_age = ten_years.groupby("age")["TenYearCHD"].count().reset_index()
group_age
```

```python
#Plot a line graph to show total number of people having a chance of coronary heart disease in 10 years

plt.plot(group_age["age"], group_age["TenYearCHD"], label = "Total number of people having a heart disease in 10 years")
plt.legend()

plt.show()


```
